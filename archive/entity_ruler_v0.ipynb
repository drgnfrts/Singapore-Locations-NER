This was a test ipynb to create a EntityRuler for a smaller sample, namely MRT stations.

```python
import spacy
from spacy.lang.en import English
from spacy.pipeline import EntityRuler
import json

def load_data(file):
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
    return (data)

def save_data(file, data):
    with open (file, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)
```

```python
data = load_data('../data/extracted_locations/extracted_mrt_stns.json')

print(len(data))
name_alternatives = []
bukit_stations = []
for item in data:
    item = item.lower()
    print(item)
    name_alternatives.append(item)
```

```python
#the damned testlist

list_of_abbreviations = [" mrt", " mrt stn", " stn", " station", ""]
final_list_of_mrt_stations = []
for stations in name_alternatives:
    final_list_of_mrt_stations.append(stations)
    for i in range(len(list_of_abbreviations)):
        replacement = stations.replace(" mrt station", list_of_abbreviations[i])
        final_list_of_mrt_stations.append(replacement)

print(final_list_of_mrt_stations)
```

```python
def create_training_data(type):
    data = final_list_of_mrt_stations
    patterns = []
    for stations in data:
        pattern = {
            "label": type,
            "pattern": stations
        }
        patterns.append(pattern)
    return(patterns)
```

```python
patterns = create_training_data("LOC")
nlp = spacy.load("en_core_web_md")
ruler = nlp.add_pipe("entity_ruler", before = "ner")
ruler.add_patterns(patterns)
nlp.to_disk("mrt_ner")
```
